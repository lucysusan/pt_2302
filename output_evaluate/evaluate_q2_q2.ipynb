```python
import matplotlib.pyplot as plt
import pandas as pd
import os
import pyfolio as pf
import warnings

warnings.filterwarnings('ignore')
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']

index_sid = '000906.SH'
data_folder = '../OUT_DATA/0301/'
file_list = os.listdir(data_folder)

print(file_pkl := [f for f in file_list if f.endswith('ret_index')])
file = [f for f in file_pkl if '2quarter_2quarter' in f][0]
```

```python

print('\n\n--------------------------------------------------------------------------\n', file)
data = pd.read_pickle(data_folder + file)
data.plot(figsize=(30, 18))
plt.title('return')
plt.show()
pf.create_full_tear_sheet(data['ret'], benchmark_rets=data[index_sid])
```
